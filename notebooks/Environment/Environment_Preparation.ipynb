```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mne
import tensorflow as tf
import sklearn
import shap
import lime
from PyQt5 import QtWidgets

print("--- 🔍 INICIANDO TESTE DE AMBIENTE ---")

# 1. Teste de Versão e Python
print(f"✅ Python {sys.version.split()[0]} detectado.")

# 2. Teste do MNE (Simulando um sinal de EEG)
try:
    info = mne.create_info(ch_names=['Fp1', 'Fp2', 'Cz'], sfreq=100, ch_types='eeg')
    data = np.random.randn(3, 100)
    raw = mne.io.RawArray(data, info, verbose=False)
    print(f"✅ MNE {mne.__version__}: Processamento de sinal funcional.")
except Exception as e:
    print(f"❌ Erro no MNE: {e}")

# 3. Teste do TensorFlow (Cálculo Matemático na CPU/GPU)
try:
    v = tf.constant([[1.0, 2.0], [3.0, 4.0]])
    _ = tf.matmul(v, v)
    gpu = tf.config.list_physical_devices('GPU')
    print(f"✅ TensorFlow {tf.__version__}: Motor de IA ativo. (GPU: {'Sim' if gpu else 'Não'})")
except Exception as e:
    print(f"❌ Erro no TensorFlow: {e}")

# 4. Teste de XAI (Explicabilidade)
try:
    _ = shap.Explanation(values=np.array([0, 1]), data=np.array([0, 1]))
    print(f"✅ SHAP & LIME: Ferramentas de explicabilidade prontas.")
except Exception as e:
    print(f"❌ Erro em XAI: {e}")

# 5. Teste de Interface Gráfica (PyQt5)
try:
    app = QtWidgets.QApplication.instance()
    if not app:
        app = QtWidgets.QApplication(sys.argv)
    print("✅ PyQt5: Interface gráfica disponível para plots interativos.")
except Exception as e:
    print(f"❌ Erro no PyQt5: {e}")

print("\n--- ✨ RESULTADO FINAL: AMBIENTE PRONTO PARA O GITHUB! ---")
```
